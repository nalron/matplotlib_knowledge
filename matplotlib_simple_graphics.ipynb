### Matplotlib : Data Visualization

At the beginning, I recommend to set the font size, because by default is small and not clear. We can use `plt.RcParams.update()`

```python
# Import it into our Python environment:
from matplotlib import pyplot as plt
# Both code:
import matplotlib.pyplot as plt
import numpy as np
```

### Basic Line Plot
Line graphs are helpful for visualizing how a variable changes over time.

```python
# Basic Line Plot: How a variable changes over time
# For instance, we can explain many things like "average prices of gasoline over the past decade"
# Diesel Price in France between January and October 2018:
plt.rcParams.update({'font.size': 13})
date = list(range(10))
price = [1.40, 1.38, 1.37, 1.40, 1.45, 1.46, 1.45, 1.45, 1.48, 1.52]

plt.plot(date, price)
plt.show()
```

```python
# Multiple line plots!
# For instance, we add a new line to show a type of tax:
date = list(range(10))
price = [1.40, 1.38, 1.37, 1.40, 1.45, 1.46, 1.45, 1.45, 1.48, 1.52]
tax = [row+0.03 for row in price]

plt.plot(date, price)
plt.plot(date, tax)
plt.show()
```

### Linestyles
Linestyles and Markers are applied to make the lines on a graph easier to understand, and to mark specific points on the graph clearly.

```python
# Linestyles:
# Keyword "color" to specify a HTML color name or a HEX code.
# Keyword "linestyle" to change make a line dotted or dashed: '--', ':'
# Keyword "marker" to add a marker: 'o', 's', '*'
plt.plot(date, price, color = 'purple', linestyle = "dashed")
plt.plot(date, tax, marker = 'o')
plt.show()
```

### Legends
When we've multiple lines on a single graph we can label them by using the command `plt.legend()`.

```python
# Legend: plt.legend()
# So useful with multiple lines…

plt.plot(date, price, color = 'purple', linestyle = "dashed")
plt.plot(date, tax, marker = 'o')

plt.legend(['Price', 'Tax'])
plt.show()
```

```python
# Legend: plt.legend()
# If we want, we can use the keyword label inside of plt.plot(). 

plt.plot(date, price, color = 'purple', linestyle = "dashed", label = 'Tax')
plt.plot(date, tax, marker = 'o', label = 'Price')

plt.legend()
plt.show()
```

### Axis and Labels
We want to get a better view (more details)with a Zoom, so we're gonna use `plt.axis()`. Use `plt.xlabel()`, `plt.ylabel()` for labelling x and y axes. `plt.title()` will use to label the Plot Title.

```python
# Axis: plt.axis([])
# We need to zoom in or out of the plot!
# This list should contain:

# The minimum x-value displayed
# The maximum x-value displayed
# The minimum y-value displayed
# The maximum y-value displayed

plt.plot(date, price, color = 'purple', linestyle = "dashed")
plt.axis([2, 8, 1.40, 1.52])
plt.show()
```

```python
# Label the Axes: plt.xlabel(), plt.ylabel() and plt.title()
# Show these plots to other people to convince them of important trends in our data.
# We’ll want to make our plots look as professional as possible.
plt.plot(date, price, color = 'purple', linestyle = "dashed")
plt.xlabel('Months')
plt.ylabel('Price')
plt.title('Diesel Price in France between January / October 2018')
plt.show()
```

### Subplots
Sometimes, we need to explain it with many different subplots. Use this command `plt.subplot()`.

```python
# Subplots: plt.subplot()
# 3 arguments into command:

# Number of rows of subplots
# Number of columns of subplots
# Index of the subplot we want to create

plt.subplot(1, 2, 1)
plt.plot(date, price, color = 'purple', linestyle = "dashed")

plt.subplot(1, 2, 2)
plt.plot(date, tax, marker = 'o')

plt.show()
```

```python
# Customize the spacing between our subplots: .subplots_adjust()
# .subplots_adjust() has some keyword arguments that can move your plots within the figure:
# left(default 0.125), right(0.9), bottom(0.1), top(0.9), wspace(0.2), hspace(0.2)

plt.subplot(1, 2, 1)
plt.plot(date, price, color = 'purple', linestyle = "dashed")

plt.subplot(1, 2, 2)
plt.plot(date, tax, marker = 'o')

plt.subplots_adjust(wspace=0.45, bottom=0.2)
plt.show()
```

### Modify Ticks
We have to create an axes object `ax` and call `plt.subplot()`like it : `ax = plt.subplot(1, 1, 1)` or `ax = plt.subplot()`.

#### Both methods for plot and axes
`plt.xticks()`
`ax.set_xticks()`

`plt.xticklabels()`
`ax.set_xticklabels()`

`plt.yticklabels()`
`ax.set_yticklabels()`

`plt.xlabel()`
`ax.set_xlabel()`

```python
# Modify Ticks:

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'nov']

date = list(range(10))
price = [1.40, 1.38, 1.37, 1.40, 1.45, 1.46, 1.45, 1.45, 1.48, 1.52]

ax = plt.subplot(1,1,1)
ax.set_xticks(date)
ax.set_xticklabels(month_names)

plt.plot(date, price, color = 'purple', linestyle = "dashed")
plt.show()
```

### Figures
We can use the command `plt.figure()` to create new figures. Add the keyword `figsize=(width, height)` to set the size of the figure. And use the command `plt.savefig()` to save out to many different file formats, such as `.png`, `.svg`, or `.pdf`.

```python
# Example with our last plot "Diesel Price in France between January / October 2018":
# So, we can add plt.figure(figsize=(8, 16))

date = list(range(10))
price = [1.40, 1.38, 1.37, 1.40, 1.45, 1.46, 1.45, 1.45, 1.48, 1.52]
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'nov']
    
plt.figure(figsize=(14, 7))
plt.plot(date, price, color = 'black', label='Diesel')

plt.xticks(date, month_names, rotation = 30)
plt.axis([0, 9, 1.35, 1.52])

plt.xlabel('Months')
plt.ylabel('Price')
plt.title('Diesel Price in France between January / October 2018')
plt.legend(loc='upper left')

plt.show()
```

### Styles
We can check quickly many styles from Matplotlib. Try with `print(plt.style.available)` 

Matplotlib is very flexible, we can customize all parameters, there is also other ways with these styles.

```python
print(plt.style.available[:10])
```

```python
fig = plt.figure(figsize=(20, 30))
for i in range(10):
    fig.add_subplot(5, 2, i+1)
    plt.style.use(plt.style.available[i])
    plt.plot(date, price, linestyle = 'dotted')
    plt.title(plt.style.available[i], color = 'blue')
```

```python
# Example fig in sinus plot:
plt.style.use('classic')
fig=plt.figure(figsize=(10,5))
ax = plt.axes(facecolor='black')
plt.grid(color='w', linestyle='solid')

x = np.linspace(0, 10, 1000)

ax.plot(x, np.sin(x), color = 'green') # Instead of plt.plot(x, np.sin(x))
ax.plot(x, np.cos(x), color = 'blue')
ax.set_xlabel('x')
ax.set_ylabel(' cos(x) / sin(x)')
ax.set_title('Example with fig structure')

plt.show()
```
